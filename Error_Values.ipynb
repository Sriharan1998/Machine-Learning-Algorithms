```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Product Name'] = le.fit_transform(df['Product Name'])
```

```python
x = df.iloc[:,:-1].values
```

```python
x
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size = 1/3, random_state=7)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mlr_reg = LinearRegression()
```

```python
mlr_reg.fit(x_train,y_train)
```

```python
y_predict = mlr_reg.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import(mean_absolute_error,mean_squared_error)
```

```python
MAE = mean_absolute_error(y_test,y_predict)
```

```python
MAE
```

```python
MSE = mean_squared_error(y_test,y_predict)
```

```python
MSE
```

```python
RMSE = np.sqrt(MSE)
lrmse  = np.log(np.sqrt(MSE))
```
